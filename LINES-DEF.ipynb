```python
import numpy as np
from scipy.interpolate import griddata
import matplotlib.pylab as plt
from matplotlib import cm
from matplotlib.cm import ScalarMappable

# generate example data
import pandas as pd

#LAB

data1 = pd.read_csv('LINE_LAB.txt', sep = '\s+', header = None)
data1 = pd.DataFrame(data1)

y1 = data1[1]-3
HmT = abs(data1[3])

#mT to T

HT = 0.001*HmT

plt.plot(y1,HT, label = 'Lab')

#COMSOL

data2 = pd.read_csv('LINE_COMSOL(1.3T).txt', sep = '\s+', header = None)
data2 = pd.DataFrame(data2)

y2 = data2[0]
H = data2[1]

plt.plot(y2,H, label = 'Comsol')
plt.xlim(-5,5)
plt.ylim(0.6,0.775)
#plt.yticks(np.arange(0.6,1.1,0.075))
plt.xticks(np.arange(-5,5,1))
plt.xlabel('Y(mm)')
plt.ylabel('H (T)')
plt.legend()
plt.show()
```
